```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
import os
```

```python
# Try loading dataset from CSV
Csv_file = '/content/naive_bayes_data.csv'
if os.path.exists(Csv_file):
    df = pd.read_csv(Csv_file)
else:
     raise FileNotFoundError("Dataset not found. please upload naive_bayes.data.csv")
```

```python
print(df.info())
```

```python
# Convert categorical feature to numeric
if df['feature1'].dtype == 'object':
    df['feature1'] = df['feature1'].astype('category').cat.codes
```

```python
# Define features and target  variable
X = df[['feature1', 'feature2']].values
Y = df['class_label'].values
```

```python
X
```

```python
Y
```

```python
# Split into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create and train the Naive Bayes model
model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluate model
accuracy = accuracy_score(Y_test, y_pred)
report = classification_report(Y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("classification Report:\n", report)
```

```python
# Function to predict class label for new data
def predict_class(feature1_value, feature2_value):
    # Convert categorical value to numeric if needed
    if isinstance(feature1_value, str):
        feature1_value = pd.Series([feature1_value]).astype('category').cat.codes[0]
    prediction = model.predict(np.array([[feature1_value, feature2_value]]))
    print(f"Predict class label: {prediction[0]}")
    return prediction[0]
```

```python
# Example prediction
predict_class('A', 9)  # change values as needed
```
